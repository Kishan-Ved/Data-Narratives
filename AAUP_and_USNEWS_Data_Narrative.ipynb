```python
# Correlation matrix and heatmap

import pandas as pd
import seaborn as sns

# Import the dataset
# data = pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/usnews.data',header=None)

for i in range(3,35):
    usnews[i] = pd.to_numeric(usnews[i], errors='coerce')
usnews['total_exp'] = pd.to_numeric(usnews['total_exp'],errors='coerce')
# Compute the correlation matrix
correlation_matrix = usnews.corr()
correlation_matrix = correlation_matrix[abs(correlation_matrix) > 0.9]

# Print the correlation matrix
# print(correlation_matrix)

# Visualize the correlations with a heatmap
plt.figure(figsize=(17,13))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Final cell - The colleges having a lower acceptance rate (difficult to get into) have a higher graduation rate

from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt
usnews['Acceptance Ratio'] = usnews['Acceptance Ratio'].fillna(0)
usnews[34] = usnews[34].fillna(0)

df = usnews.loc[:,['Acceptance Ratio',34]].copy()
df = df[(df!=0).all(1)]

x = df['Acceptance Ratio']
y = df[34]

x = x.T
m = np.c_[np.ones(len(x)),x]
# print(m)
p = linalg.lstsq(m,y)[0]
# print(p)
a = p[0]
b = p[1]
print("a is :",a,"\nb is :",b)
xx = np.arange(0,1,0.001)
yy = a + b*xx
plt.plot(x,y,'o')
plt.plot(xx,yy)
plt.xlabel("Acceptance Ratio")
plt.xlim([None,1])
plt.ylim([None,100])
plt.ylabel("Graduation rate")
plt.title("Trend between acceptance ratio and graduation rate")
plt.show()

# Correlation matrix and heatmap

import pandas as pd
import seaborn as sns

# Import the dataset
# data = pd.read_csv('http://lib.stat.cmu.edu/datasets/colleges/usnews.data',header=None)

for i in range(4,17):
    aaup[i] = pd.to_numeric(aaup[i], errors='coerce')
# Compute the correlation matrix
correlation_matrix = aaup.corr()
# correlation_matrix = correlation_matrix[abs(correlation_matrix) > 0.9]

# Print the correlation matrix
# print(correlation_matrix)

# Visualize the correlations with a heatmap
plt.figure(figsize=(10,7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()

usnews['Average Salary'] = aaup[4]*100
usnews['Average Salary'] = usnews['Average Salary'].fillna(0)
usnews

# Final Cell

from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt

usnews['Average Salary'] = aaup[4]*100
usnews['Average Salary'] = usnews['Average Salary'].fillna(0)
usnews['Total Faculty'] = aaup[16].replace('*',0)
usnews[29] = usnews[29].fillna(0)
usnews['No of PhD'] = usnews['Total Faculty']*usnews[29]/100
df = usnews.loc[:,['Average Salary','No of PhD']].copy()
df = df[(df!=0).all(1)]
y = df['Average Salary']
x = df['No of PhD']

x = x.T
m = np.c_[np.ones(len(x)),x]
# print(m)
p = linalg.lstsq(m,y)[0]
# print(p)
a = p[0]
b = p[1]
print("a is :",a,"\nb is :",b)
xx = np.arange(0,2000,1)
yy = a + b*xx
plt.plot(x,y,'o')
plt.plot(xx,yy,color='orange')
plt.xlabel("Number of Ph.D Faculty")
plt.xlim([0,1250])
# plt.ylim([2,None])
plt.ylabel("Average Salary")
plt.title("Trend between Number of Ph.D. faculty and Averge Salary")
plt.show()
```
